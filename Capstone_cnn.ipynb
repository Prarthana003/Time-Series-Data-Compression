```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Load the dataset
data = pd.read_csv('periodic.tsv', sep='\t', header=None)

# Separate the features and labels
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values

# Normalize the features
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Check the unique period lengths
unique_periods = np.unique(y)
print("Unique periods:", unique_periods)
num_classes = len(unique_periods)
print("Number of classes:", num_classes)

# Adjust labels to be zero-indexed for to_categorical
label_mapping = {period: index for index, period in enumerate(unique_periods)}
y_mapped = np.array([label_mapping[period] for period in y])

# Convert labels to categorical
y_categorical = to_categorical(y_mapped, num_classes=num_classes)

# Reshape X to fit the CNN input shape (samples, height, width, channels)
X = X.reshape((X.shape[0], X.shape[1], 1, 1))

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.2, random_state=42)

```

```python
# Define the CNN model
model = Sequential([
    Conv2D(32, kernel_size=(3, 1), activation='relu', input_shape=(X_train.shape[1], 1, 1)),
    MaxPooling2D(pool_size=(2, 1)),
    Conv2D(64, kernel_size=(3, 1), activation='relu'),
    MaxPooling2D(pool_size=(2, 1)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')  # Output layer for classification
])
```

```python

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', 'mae'])

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the model
test_loss, test_accuracy, test_mae = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_accuracy}')
print(f'Test MAE: {test_mae}')
```
